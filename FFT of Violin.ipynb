```python
import os
import matplotlib.pyplot as plt
import librosa, librosa.display
import IPython.display as ipd
import numpy as np
```

```python
BASE_FOLDER = "/Users/Praveena Acharya/Desktop/Audio Spectral Analysis/Audio Data"
sound_file = "violin-C4.wav"
sound_file1 = "violin-C5.wav"
sound_file2 = "violin-C6.wav"
# C4 = 261Hz
# C5 = 523Hz
# C6 = 1046Hz
```

```python
ipd.Audio(os.path.join(BASE_FOLDER, sound_file))
```

```python
ipd.Audio(os.path.join(BASE_FOLDER, sound_file1))
```

```python
ipd.Audio(os.path.join(BASE_FOLDER, sound_file2))
```

```python
# load sounds
voice, sr = librosa.load(os.path.join(BASE_FOLDER, sound_file))
voice1, _ = librosa.load(os.path.join(BASE_FOLDER, sound_file1))
voice2, _ = librosa.load(os.path.join(BASE_FOLDER, sound_file2))
```

```python
f_ratio = 0.05
```

```python
plt.figure(figsize=(20, 5))
plt.plot(voice)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time Domain')
```

```python
X = np.fft.fft(voice)
X_mag = np.absolute(X)
f = np.linspace(0, sr, len(X_mag))
f_bins = int(len(X_mag)*0.02)
plt.figure(figsize=(20, 5))
plt.plot(f[:f_bins], X_mag[:f_bins])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency Domain - Fourier Transform ')
plt.show()
```
